```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
data=pd.read_csv(r"C:\Users\RAHUL KRISHNA\Desktop\Weather Prediction Data\Cleaning & Preprocessing\weatherml.csv")
```

```python
data
```

```python
data.drop(['Cloud Cover'],axis=1,inplace=True)
```

```python
data
```

```python
x=data[['Summary_encoded','precip_encoded','daily_encoded']]
y=data['Daily Summary']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
data
```

```python
log=LogisticRegression()                                            # Logistic Regression Model-train-0.1
```

```python
log.fit(x_train,y_train)                               
```

```python
log.score(x_train,y_train)
```

```python
log.score(x_test,y_test)
```

```python
nb=MultinomialNB()                                                             # Naive_Bayes Model-train(0.2)
```

```python
nb.fit(x_train,y_train)
```

```python
nb.score(x_train,y_train)
```

```python
nb.score(x_test,y_test)
```

```python
for c in range(1,50,5):                                                       #KNN Neighbors Model-train-(0.2737)
    kn=KNeighborsClassifier(n_neighbors=c) 
    kn.fit(x_train,y_train)
    print(kn.score(x_train,y_train))
    print(c,kn.score(x_test,y_test))
```

```python
rf=RandomForestClassifier()                                          # Random Forest Classifier-train(0.2891)
```

```python
rf.fit(x_train,y_train)
```

```python
rf.score(x_train,y_train)                                               
```

```python
rf.score(x_test,y_test)
```

```python
import pickle
with open('model.pickle','wb')as model:

    pickle.dump(rf,model)
```
